```python
import glob
import os
import cv2
from shutil import copyfile
```

```python
directories = os.listdir("PCBData")
annotations = []
for directory in directories:
    anns = glob.glob('PCBData\\'+directory+'\\'+directory.split('group')[-1]+'_not\\*')
    imgs = glob.glob('PCBData\\'+directory+'\\'+directory.split('group')[-1]+'\\*')

    for a in anns:
        with open(a, "r") as text_file:
            data = text_file.read().splitlines()
        temp = a.replace('_not', '').replace('.txt', '_temp.jpg')
        test = a.replace('_not', '').replace('.txt', '_test.jpg')
        temp_img = cv2.imread(temp)
        test_img = cv2.imread(test)

        for i, annotation in enumerate(data):
            a = annotation.split(' ')
            x1 = int(a[0])
            y1 = int(a[1])
            x2 = int(a[2])
            y2 = int(a[3])
            category = int(a[4])
            temp_img_crop = temp_img[y1:y2, x1:x2]
            test_img_crop = test_img[y1:y2, x1:x2]
            # print(a)
            # print(temp_img_crop.shape)

            cv2.imwrite('PCBData_convtd\\og\\'+str(category)+'\\'+temp.split('\\')[-1].split('.')[0]+'_'+str(i)+'.jpg', temp_img_crop)
            cv2.imwrite('PCBData_convtd\\test\\'+str(category)+'\\'+test.split('\\')[-1].split('.')[0]+'_'+str(i)+'.jpg', test_img_crop)
        annotations += (data)
file=open('PCBData_convtd\\annotations.txt','w')
for items in annotations:
    file.writelines(items+'\n')
file.close()
```
